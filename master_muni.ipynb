```python
# https://www.ine.es/daco/daco42/codmun/diccionario22.xlsx
# https://www.ine.es/daco/daco42/codmun/codmun02/02codmunmapa.htm
# https://www.ine.es/daco/daco42/clasificaciones/codccaa.xls
```

```python
import pandas as pd
```

```python
df_muni = pd.read_excel("diccionario22.xlsx", header = 1)
df_muni.columns = df_muni.columns.str.lower()
df_muni.columns = df_muni.columns.str.strip()
df_muni["cod_ccaa"] = df_muni["codauto"].apply(str).str.zfill(2)
df_muni["cod_prov"] = df_muni["cpro"].apply(str).str.zfill(2)
df_muni["cmun"] = df_muni["cmun"].apply(str).str.zfill(3)
df_muni["cod_muni"] = df_muni["cod_prov"].astype(str) + df_muni["cmun"].astype(str) 
df_muni.drop(columns=["dc"], inplace = True)
df_muni = df_muni[["cod_ccaa","cod_prov","cod_muni","nombre"]].copy()
df_muni.rename(columns = {'nombre':'municipio'}, inplace = True)
print(df_muni.shape)
df_muni.head()
```

```python
df_prov = pd.read_excel("codprov.xls", header = 1)
df_prov.columns = df_prov.columns.str.lower()
df_prov.columns = df_prov.columns.str.strip()
df_prov.rename(columns = {'codigo':'cod_prov', 'literal':'provincia'}, inplace = True)
df_prov["cod_prov"] = df_prov["cod_prov"].apply(str).str.zfill(2)
print(df_prov.shape)
df_prov.head()
```

```python
df_ccaa = pd.read_excel("codccaa.xls", header = 1)
df_ccaa.columns = df_ccaa.columns.str.lower()
df_ccaa.columns = df_ccaa.columns.str.strip()
df_ccaa.rename(columns = {'codigo':'cod_ccaa', 'literal':'ccaa'}, inplace = True)
df_ccaa["cod_ccaa"] = df_ccaa["cod_ccaa"].apply(str).str.zfill(2)
print(df_ccaa.shape)
df_ccaa.head()
```

```python
print(df_muni.shape)
print(df_prov.shape)
df = pd.merge(df_muni, df_prov, on = "cod_prov", how = "left")
print(df.shape)
df.head()
```

```python
print(df.shape)
print(df_ccaa.shape)
df = pd.merge(df, df_ccaa, on = "cod_ccaa", how = "left")
print(df.shape)
df.sample(2)
```

```python
df = df[["cod_muni","municipio","cod_prov","provincia","cod_ccaa","ccaa"]]
print(df.shape)
df.sample(10)
```

```python
df.to_csv("master_muni.csv", sep = "|", index=False)
```
